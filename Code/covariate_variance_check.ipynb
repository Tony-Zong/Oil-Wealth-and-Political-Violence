```python
covariate_df = df[['year', 'code3', 'numcode', 'pop_maddison', 'religion_fractionalization', 
                   'logpopdens', 'logGDP_M', 'ecgrowth',
                   'democracy', 'x_dem', 'ethnic_fractionalization', 'language_fractionalization',
                   'efrac', 'lfrac', 'rfrac', 'ETHPOL_reynal', 'ETHFRAC_reynal', 'RELPOL_reynal', 'RELFRAC_reynal',
                   'mountain', 'logmountain', 'leg_british', 'rule_law96', 'land_area', 'smallregion', 
                   'largeregion', 'out_regdisaster', 'logoutreg',  'lowincsample']]
```

```python
unique_countries = df['code3'].unique()
```

```python
for col in covariate_df.columns:
    if col == 'code3':
        continue
    print('Column Name:', col)
    print('Number of Unique Values in Col:', len(covariate_df[col].unique()))
    print('Number of Missing Values in Col:', covariate_df[col].isnull().sum())
    print('Mean of Col:', round(covariate_df[col].mean(), 5))
    print('Standard Deviation of Col:', round(covariate_df[col].std(), 5))
    print('\n')
```

```python
for col in covariate_df.columns:
    if col == 'code3':
        continue
    total_variance = 0
    total_countries = 0
    for country in unique_countries:
        if np.isnan(covariate_df[covariate_df['code3'] == country][col].dropna().std()):
            continue
        total_variance += covariate_df[covariate_df['code3'] == country][col].dropna().std()
        total_countries += 1

    print('Column:', col)
    print('Avg Std Across Countries:', total_variance / total_countries)
    print('\n')
```

```python
for country in unique_countries:
    print(covariate_df[covariate_df['code3'] == country].describe())
```
